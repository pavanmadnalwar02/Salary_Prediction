# Data Preprocessing

```python
data.info()
```

```python
data['Job Title'].nunique()
```

```python
data.isnull().sum()
```

```python
#drop missing value
data.dropna(inplace=True)
data
```

**Encoding**


```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
# Encode 'Gender', 'Education Level', 'Job Title'
data['Gender'] = lb.fit_transform(data['Gender'])
data['Education Level'] = lb.fit_transform(data['Education Level'])
data['Job Title'] = lb.fit_transform(data['Job Title'])
```

```python
data
```

```python
X=data.drop('Salary',axis=1)
Y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=42)
```

# Linear Regression

```python
#Modelling
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
#prediction
Ypred_train=model.predict(X_train)
#evaluation
from sklearn.metrics import r2_score
print('Train Accuracy:',r2_score(Y_train,Ypred_train))
#CrossVaalidaion Score
from sklearn.model_selection import cross_val_score
print('CV score:',cross_val_score(model,X_train,Y_train,cv=5,scoring='r2').mean())
#predication of test data
Ypred_test=model.predict(X_test)
#Evaluation of test data
print('test_r2:',r2_score(Y_test,Ypred_test))
```

```python
%%timeit
model.fit(X_train,Y_train)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_converter=PolynomialFeatures(degree=1)
X_train_poly=pd.DataFrame(polynomial_converter.fit_transform(X_train))
#Modelling
from sklearn.linear_model import LinearRegression
poly_model=LinearRegression()
poly_model.fit(X_train_poly,Y_train)
#Prediction
Ypred_train=poly_model.predict(X_train_poly)
#Evaluation
from sklearn.metrics import r2_score

print('Train Accuracy:',r2_score(Y_train,Ypred_train))
#Cross_Validation
from sklearn.model_selection import cross_val_score
print('CV score:',cross_val_score(poly_model,X_train_poly,Y_train,cv=5,scoring='r2').mean())
#test Data
X_test_poly=pd.DataFrame(polynomial_converter.transform(X_test))
Ypred_test=poly_model.predict(X_test_poly)
print('Test Accuracy:',r2_score(Y_test,Ypred_test))
```

```python
%%timeit
poly_model.fit(X_train,Y_train)
```

# Lasso Regression

```python
#Hyperparameter tunning for Lasso Regression
from sklearn.linear_model import Lasso
esitmator=Lasso()
from sklearn.model_selection import GridSearchCV
param_grid={'alpha':list(range(1,100))}
lasso_gd= GridSearchCV(esitmator,param_grid,scoring='r2',cv=5)
lasso_gd.fit(X_train,Y_train)
lasso_gd.best_estimator_
```

```python
#Modelling
from sklearn.linear_model import Lasso
lasso_model=Lasso(alpha=1)
lasso_model.fit(X_train,Y_train)

#Prediction
Ypred_train=lasso_model.predict(X_train)
#Evaluation
from sklearn.metrics import r2_score
print('Train accuracy:',r2_score(Y_train,Ypred_train))
#CrossValidation
from sklearn.model_selection import cross_val_score
print('cv score;',cross_val_score(lasso_model,X_train,Y_train,cv=5,scoring='r2').mean())
#test data
Ypred_test=lasso_model.predict(X_test)
print('test accuracy:',r2_score(Y_test,Ypred_test))
```

```python
%%timeit
lasso_model.fit(X_train,Y_train)
```
